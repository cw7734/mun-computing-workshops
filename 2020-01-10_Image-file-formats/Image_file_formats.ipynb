```python
import sys
import math
import numpy as np
import matplotlib.pyplot as plt
%cd "2020-01-10_Image-file-formats"
```

```python
# our function:
def gauss( x ):
    "gaussian distribution"
    sig = math.sqrt( 0.2 )
    mu = 0.0

    a = 1 / ( sig * math.sqrt( 2 * math.pi ) )
    b = mu
    c = sig

    return 1 / ( sig * math.sqrt( 2 * math.pi ) ) * math.pow( math.e, ( -1 * ( x - b ) ** 2 / ( 2 * c ** 2 ) ) )
```

```python
# generate the plot:
t = np.arange( -4.0 , 4 , 0.05 )
s = []
for x in t:
    s.append( gauss( x ) )

fig = plt.figure( 1, figsize=( 6.0, 4.0 ), dpi=100 , linewidth=0.25 )
# ax = fig.add_subplot( 111 )
# ax.set_frame_on( "False" )
plt.plot( t, s, 'b-', linewidth=1.0 )
plt.grid( True , linewidth=0.5 , marker=3 )
plt.axis( [-4.0 , 4.0, -0.01, 1.0] )
plt.tick_params( 'both', direction='in', length=3, width=0.75 )

plt.show()
```

```python
# generate plot at 4 resolutions for 3 file formats
def genplot(fname, dpi=100, fmt="png", limit=4):
    "save the plots"
    t = np.arange( limit * -1.0 , limit , 0.05 )
    s = []
    for x in t:
        s.append( gauss( x ) )

    fig = plt.figure( 1, figsize=( 6.0, 4.0 ), dpi=dpi , linewidth=0.25 )
    plt.plot( t, s, 'b-', linewidth=1.0 )
    plt.grid( True , linewidth=0.5 , marker=3 )
    plt.axis( [limit * -1.0 , limit, -0.01, 1.0] )
    plt.tick_params( 'both', direction='in', length=3, width=0.75 )

    plt.savefig(fname, dpi=dpi)

for dpi in [72, 100, 300, 600]:
    for fmt in ['png', 'svg', 'pdf']:
        fname = "gaussian/gaussian_{dpi:03d}dpi.{fmt:}".format(dpi=dpi, fmt=fmt)
        genplot(fname,  dpi=dpi, fmt=fmt, limit=4,)
```
